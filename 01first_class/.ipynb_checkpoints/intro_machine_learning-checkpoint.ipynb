# Acquisition and exploratory analysis of COVID data from brasil.io for the state of Piauí
### Through the dataset cases_covid19.csv, the names of the 224 cities in Piauí were captured. However, these names could also be taken from the IBGE website. It is important that the most up-to-date data is captured on the website: https://brasil.io/home/. After capturing the city name, we will use openstreetmap to capture latitude and logitude;

## Imports

```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
```

## Data acquisition

```python
data = pd.read_csv('caso.csv')
```

```python
data.head()
```

```python
# dataset size
# first number is the number of lines
# second number is the number of columns
data.shape
```

```python
# selecting in the dataset only the data from piauí by the 'state' column
data_piaui = data[data['state'] == 'PI']
```

```python
data_piaui.head()
```

```python
# basic information about the dataset
data_piaui.info()
```

```python
# selecting only cities with column 'place_type'
cidades = data_piaui[data_piaui['place_type'] == 'city']
```

```python
# unique() selects unique data, for example:
# if you have n cities with the same name, only one will be saved
list_cities = cidades['city'].unique()
```

```python
# viewing the cities
list_cities
```

```python
# array size
len(list_cities)
```

## Capturing latitudes and longitudes with openstreetmaps

```python
latitudes = []
longitudes = []
for city in list_cities:
    request = 'https://nominatim.openstreetmap.org/search?city='+city+'&state=Piaui&format=json'
    response = requests.get(request)
    response_data = response.json()[0]
    
    latitude = float(response_data["lat"])
    longitude = float(response_data["lon"])
    latitudes.append(latitude)
    longitudes.append(longitude)
```
